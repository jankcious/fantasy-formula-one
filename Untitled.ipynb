```python
import numpy as np
import pandas as pd
```

```python
rolls = np.random.random_integers(1, 20, 10000000)
rolls2 = np.random.random_integers(1, 20, 10000000)
matrix = pd.DataFrame(rolls)
```

```python
matrix['roll2'] = rolls2
```

```python
matrix
```

```python
matrix.columns = ['roll1', 'roll2']
matrix
```

```python
%timeit matrix[(matrix['roll1'] == 19) | (matrix['roll1'] == 20) | (matrix['roll2'] == 19) | (matrix['roll2'] == 20) | \
       (matrix['roll1'] == matrix['roll2'])].count() / 10000000 * 100
```
